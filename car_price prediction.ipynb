```python
final_data=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_data.head()
```

```python
final_data['Current_year']=2020
```

```python
final_data.head()
```

```python
final_data['no_year']=final_data['Current_year']-final_data['Year']
```

```python
final_data.head()
```

```python
final_data.drop(['Year'],axis=1,inplace=True)
```

```python
final_data.drop(['Current_year'],axis=1,inplace=True)
```

```python
final_data.head()
```

```python
final_data=pd.get_dummies(final_data,drop_first=True)
```

```python
final_data.head()
```

```python
sns.pairplot(final_data)
```

```python
corr=final_data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,)
```

```python
X=final_data.iloc[:,1:]
y=final_data.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
# MODEL1 LINEAR REGRESSION
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error

lir=LinearRegression()
lir.fit(X_train,y_train)

train_preds1=lir.predict(X_train)
test_preds1=lir.predict(X_test)
```

```python
# MODEL1 PREDICTIONS
print("mean_absolute_error on train data:",mean_absolute_error(y_train,train_preds1))
print("mean_absolute_error on test data:",mean_absolute_error(y_test,test_preds1))
```

```python
# MODEL2 DECISION TREE REGRESSOR 
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import RandomizedSearchCV

dtc=DecisionTreeRegressor()
dtc.fit(X_train,y_train)

train_preds2=dtc.predict(X_train)
test_preds2=dtc.predict(X_test)

print("mean_absolute_error on train data:",mean_absolute_error(y_train,train_preds2))
print("mean_absolute_error on test data:",mean_absolute_error(y_test,test_preds2))
```

```python
# MODEL3 KNN REGRESSOR
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=3,algorithm="brute",weights="distance")
knn.fit(X_train,y_train)

train_preds3=knn.predict(X_train)
test_preds3=knn.predict(X_test)

print("mean_absolute_error on train data:",mean_absolute_error(y_train,train_preds3))
print("mean_absolute_error on test data:",mean_absolute_error(y_test,test_preds3))
```
